### Solution 1-

<span style = 'font-size:0.8em;'>
    
Seaborn is a powerful visualization library in Python built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Here are five types of plots you can create using Seaborn, along with their uses:

1. **Scatter Plot (`sns.scatterplot`)**
   - **Use:** A scatter plot displays values for typically two variables for a set of data. It is useful for visualizing the relationship between two continuous variables.
   - **Example Use:** Exploring the correlation between height and weight.

2. **Bar Plot (`sns.barplot`)**
   - **Use:** A bar plot presents categorical data with rectangular bars. The length of the bars represents the value of the variable. It is useful for comparing the values of different groups or categories.
   - **Example Use:** Comparing the average test scores across different classes or categories.

3. **Box Plot (`sns.boxplot`)**
   - **Use:** A box plot displays the distribution of a dataset based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It is useful for identifying outliers and understanding the spread and skewness of the data.
   - **Example Use:** Analyzing the distribution of salaries within different departments.

4. **Heatmap (`sns.heatmap`)**
   - **Use:** A heatmap displays data in matrix form, where individual values are represented by color gradients. It is useful for visualizing matrices, correlations, and for finding patterns in data.
   - **Example Use:** Displaying the correlation matrix of multiple variables to find how strongly variables are related to each other.

5. **Pair Plot (`sns.pairplot`)**
   - **Use:** A pair plot shows pairwise relationships in a dataset. It creates a matrix of scatter plots for each pair of numerical variables, along with histograms or KDE plots for individual variables. It is useful for exploring relationships between multiple variables at once.
   - **Example Use:** Visualizing the relationships between all combinations of multiple features in a dataset, such as the Iris dataset features.

Each of these plots can be customized and enhanced with additional features in Seaborn to create informative and aesthetically pleasing visualizations for data analysis.
</span>

```python
# 1. Scatter Plot
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
tips = sns.load_dataset("tips")

# Scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
# 2. Bar Plot
# Sample data
tips = sns.load_dataset("tips")

# Bar plot
sns.barplot(data=tips, x="day", y="total_bill", estimator=sum)
plt.title("Total Bill by Day")
plt.show()
```

```python
# 3.Box Plot

# Sample data
tips = sns.load_dataset("tips")

# Box plot
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

```python
# 4. Heatmap
# Sample data
flights = sns.load_dataset("flights")

# Pivot the data to get a matrix format
flights_pivot = flights.pivot("month", "year", "passengers")

# Heatmap
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Heatmap of Monthly Passengers Over Years")
plt.show()
```

```python
# 5.Pair Plot
# Sample data
iris = sns.load_dataset("iris")

# Pair plot
sns.pairplot(iris, hue="species")
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
# 'pclass', y = 'age' and y = 'fare'.
```

### Solution 3-

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Box plot for Age
sns.boxplot(data=titanic, x="pclass", y="age", ax=axes[0])
axes[0].set_title("Box Plot of Age by Pclass")

# Box plot for Fare
sns.boxplot(data=titanic, x="pclass", y="fare", ax=axes[1])
axes[1].set_title("Box Plot of Fare by Pclass")

# Show the plots
plt.tight_layout()
plt.show()
```
