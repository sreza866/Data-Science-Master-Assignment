```python
# Q5. What is Distplot? Using Plotly express, plot a distplot.
```

### Solution 5-

<span style = 'font-size:0.8em;'>
    
A distplot, or distribution plot, is a type of plot that shows the distribution of a dataset. It combines elements of a histogram and a density plot to provide insights into the underlying distribution of the data. In Plotly express, you can create a similar plot using the `histogram` function with additional parameters to display both the histogram and the kernel density estimate (KDE).

While Plotly express doesn't have a direct `distplot` function (as `distplot` is a function from seaborn), you can achieve a similar result by using the `histogram` function with the `marginal` and `histnorm` parameters.
</span>


```python
tips = px.data.tips()

# Create a distribution plot for the 'total_bill' column
fig = px.histogram(tips, x="total_bill", 
                   histnorm='density',# Normalize the histogram
                   color_discrete_sequence = ["#CB391A"],
                   title="Distribution Plot of Total Bill")
fig.show("png")
```
